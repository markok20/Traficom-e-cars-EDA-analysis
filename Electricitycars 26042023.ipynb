### New dataframe

```python

writer = pd.ExcelWriter('tunnusluvut.xlsx', engine = 'xlsxwriter')


columns = ['ajoneuvoluokka', 'kayttovoima', 'merkkiSelvakielinen', 'mallimerkinta', 'kaupallinenNimi',
        'kunta', 'kayttovoima_Electricity']
df1 = pd.DataFrame(df, columns = columns)


writer.save()
```

```python
df1
```

### Adding new column electricity car

```python
df1['electricitycar'] = df['kayttovoima_Electricity'].replace({1 : 'Electricity', 0 : ''})
df1
```

### Deleting rows where electricity car is missing

```python

df1 = df1.dropna(subset=['kayttovoima_Electricity'])
df1
```

```python
df1.shape
```

```python

df1 = df1.loc[df1["kayttovoima_Electricity"] != 0]
```

```python
df1.shape
```

```python
df1
```

### The most popular electricity car models in 2021 and 2022 in Finland

```python
df_suositut = pd.crosstab(df1['mallimerkinta'], 'Total').nlargest(n = 20, columns = 'Total' )
df_suositut
```
